```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../loan_approval_data.csv")
```

handle missing values 



```python
categorical_colms = df.select_dtypes(include = ["object"]).columns
numerical_colms = df.select_dtypes(include = ["float64"]).columns


from sklearn.impute import SimpleImputer

cat_imp = SimpleImputer(strategy="most_frequent")
df[categorical_colms] = cat_imp.fit_transform(df[categorical_colms])

num_imp = SimpleImputer(strategy= "mean")
df[numerical_colms] = num_imp.fit_transform(df[numerical_colms])
```
